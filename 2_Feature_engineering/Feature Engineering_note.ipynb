# 2.3 feature selection
`class sklearn.feature_selection.SelectFromModel(estimator, threshold=None, prefit=False)`



```python
from sklearn.feature_selection import SelectFromModel
```

```python
modelData = pd.get_dummies(data=df, columns=['month', 'day', 'DC_bi'])
modelData.head()
```

```python
# 先剔除掉目标变量
# 0 for rows and 1 for columns.
# 如果是在modelData上直接删除的话，令inplace=True
xdata = modelData.drop('area', axis = 1)
ydata = modelData['area']
```

```python
xdata.head()
```

##  L1-based feature selection

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(xdata.fillna(-999), ydata)
```

```python
lasso.coef_
```

从上面的结果我们能够看到每个feature对应的系数，而那些系数为0的feature说明不重要，我们可以删掉。这里要注意一点，**不要删掉OneHotEncoder**的feature。这样的feature使我们认为做得，而且大部分为0.如果删了的话，你就把month,day全部删掉了。

只要是L1的模型，都可以做这样的操作。

当然，人工可以删没有用的feature。但我们也可以直接用sklearn来自动删除。
其实解释把模型和评估的方式一起传入。

SelectFromModel is a meta-transformer that can be used along with any estimator that has a coef_ or feature_importances_ attribute after fitting. The features are considered unimportant and removed, if the corresponding coef_ or feature_importances_ values are below the provided threshold parameter. Apart from specifying the threshold numerically, there are built-in heuristics for finding a threshold using a string argument. 

```python
# 直接传模型
model = SelectFromModel(lasso)
```

```python
# 这里transform没什么意义，为了pipeline
xdata.replace(np.inf, np.nan)
xdata.fillna(-999)
model.fit_transform(xdata)
```

# 降维方法：PCA VS TSVD

PCA应用最多的地方实在图像的压缩上.
这两个算法的区别在于：
1. PCA要计算协方差矩阵（covariance matrix)
2. TSVD不需要计算协方差矩阵

```python
from sklearn.decomposition import PCA, TruncatedSVD
pca = PCA(20)
```

```python
# 先做标准化，在做pca. 因为涉及到了矩阵计算，协方差的计算，量纲不一致，肯定有问题
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
pca.fit_transform(ss.fit_transform(Xdata)).shape
```

```python
np.sum(pca.explained_variance_ratio_)
# 说明我们选取的20个主要成分能解释91.6%的方差，能解释92%的数据变动
```

## TSVD

```python
tsvd = TruncatedSVD(20)
```

```python
tsvd.fit_transform(ss.fit_transform(Xdata))
```
